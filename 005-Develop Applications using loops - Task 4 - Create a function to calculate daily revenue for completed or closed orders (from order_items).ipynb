This program is to calculate the daily revenue of completed or closed orders. So orders need to be processed by Date to get revenue for everyday. The status can be obtained from order file and the revenue for the order can be calculated from Order items file

```python
# function to read through a file and split the order lines
def readData(filepath):
    file = open(filepath)
    data = file.read()
    datalist = data.splitlines()
    return datalist
```

```python
# function to filter completed and closed orders from order file
def filterCompletedorders(orders):
    completedOrders = []
    for i in orders:
        if(i.split(',')[3] in ('COMPLETE','CLOSED')):
            completedOrders.append(i)
    return completedOrders
        
```

```python
def getOrderidanddateindict(orders):
    orderIdanddatedict = {} # create an empty Dictionary
    for i in orders:
        # Dictionary key is the first field of orders record, 
        # Dictionary value is the second field of orders record
        # first field in orders record is Order ID
        # second field in orders record is Date of the order
        orderIdanddatedict[int(i.split(',')[0])] = i.split(',')[1]
    return orderIdanddatedict
```

```python
def getDailyRevenue(orderIdAndDateDict, orderItems):
    
    dailyRevenue = {}    # create an empty Dictionary
    for i in orderItems:
        # create a tuple with order id and total revenue of the order item
        orderIdandRevenue = (int(i.split(',')[1]),float(i.split(',')[4]))
        # check if the above tuple's order id is in the OrderIdandDateDict
        if(orderIdAndDateDict.get(orderIdandRevenue[0])):
            # if orderid is found, check if the date of the corresponding order id 
            # is in dailyRevenue dictionary
            if(dailyRevenue.get(orderIdAndDateDict[orderIdandRevenue[0]])):
                #if the date is found, it means you have id in complete status and also have 
                # an entry for that date in the dictionary, so add the revenue with the 
                # existing value
                dailyRevenue[orderIdAndDateDict[orderIdandRevenue[0]]] += orderIdandRevenue[1]
            else:
                # if the date is not found, it means this is the first entry, 
                # so create an entry in the dictionary
                dailyRevenue[orderIdAndDateDict[orderIdandRevenue[0]]] = orderIdandRevenue[1]
    return dailyRevenue
    
```

```python
orderPath = "/data/retail_db/orders/part-00000"
orders = readData(orderPath)
orderItemsPath = "/data/retail_db/order_items/part-00000"
orderItems = readData(orderItemsPath)
print(orderItems[0:5])
```

```python
completedOrders = filterCompletedorders(orders)
print(completedOrders[0:5])
```

```python
orderidanddateinDict = getOrderidanddateindict(completedOrders)
print(orderidanddateinDict[1])
```

```python
dailyRev = getDailyRevenue(orderidanddateinDict,orderItems)
print(dailyRev['2013-07-25 00:00:00.0'])
```
