This program is to extract order id and order item subtotal

```python
# function to read through a file and split the order lines
def readData(filepath):
    file = open(filepath)
    data = file.read()
    datalist = data.splitlines()
    return datalist
```

```python
orderItems = readData("/data/retail_db/order_items/part-00000")
orderItems[0:5]
```

```python
def getOrderIDandTotalPrice(OrderItems):
    orderIDandTotalPrice = []
    for items in OrderItems:
        orderIDandTotalPrice.append((items.split(',')[1],items.split(',')[4]))
    return orderIDandTotalPrice
```

```python
getOrderIDandTotalPrice(orderItems)
```
