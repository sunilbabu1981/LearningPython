Order Items whose Total price is not equal to quantity multiplied by unit price

```python
# function to read through a file and split the order lines
def readData(filepath):
    file = open(filepath)
    data = file.read()
    datalist = data.splitlines()
    return datalist
```

```python
orderItems = readData("/data/retail_db/order_items/part-00000")
print(orderItems[0:5])
```

```python
def getOrdersSubtotalNEQtymultUniPrice(orderDetails):
    OrderItemMessedDict = {}
    for Items in orderDetails:
        SubTotalwithQtyandPrice = (Items.split(',')[1],float(Items.split(',')[4]),int(Items.split(',')[3]),float(Items.split(',')[5]))
        #print("OrderID:",SubTotalwithQtyandPrice[0])
        #print("subtotal:",SubTotalwithQtyandPrice[1])
        #print("quantity:",SubTotalwithQtyandPrice[2])
        #print("unit price:",SubTotalwithQtyandPrice[3])
        #if(SubTotalwithQtyandPrice[1] != round((SubTotalwithQtyandPrice[2] * SubTotalwithQtyandPrice[3]),2)):
        if(SubTotalwithQtyandPrice[1] != (SubTotalwithQtyandPrice[2] * SubTotalwithQtyandPrice[3])):
            #print("Messedup",(SubTotalwithQtyandPrice[2] * SubTotalwithQtyandPrice[3]))
            if(OrderItemMessedDict.get(SubTotalwithQtyandPrice)):
                OrderItemMessedDict[SubTotalwithQtyandPrice] += 1
            else:
                OrderItemMessedDict[SubTotalwithQtyandPrice] = 1
    return OrderItemMessedDict
```

```python
MessedupOrderItems = getOrdersSubtotalNEQtymultUniPrice(orderItems)
MessedupOrderItems
# The mismatch is due to rounding off. When rounded up there are no rows that get mismatched 
# between the subtotal vs (quanty * unit price)
```
