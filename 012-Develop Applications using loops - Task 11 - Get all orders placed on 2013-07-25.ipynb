This program is to find all orders that were placed on 2013-07-25

```python
# function to read through a file and split the order lines
def readData(filepath):
    file = open(filepath)
    data = file.read()
    datalist = data.splitlines()
    return datalist
```

```python
orderPath = "/data/retail_db/orders/part-00000"
orders = readData(orderPath)
```

```python
def getOrdersplacedonagivendate(orders,orderdate):
    ordersfortheday = []
    for order in orders:
        if(str(order.split(',')[1][0:10]) == orderdate):
            ordersfortheday.append(order)
    return ordersfortheday
```

```python
getOrdersplacedonagivendate(orders,'2013-07-25')
```
