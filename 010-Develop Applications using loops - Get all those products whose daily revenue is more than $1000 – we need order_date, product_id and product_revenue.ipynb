Get all those products whose daily revenue is more than $1000 â€“ we need order_date, product_id and product_revenue


Thinking about the logic, we need both orders and order items to retrieve this information, so we need to sum the revenue 
of each product per day/date. Here the uniqueness is date and product. So try to club them into one string and use a dictionary to store this as key and the total revenue as value

Step 1 : Get orderid and date in a dictionary
Step 2 : for each of the matching order from orderitem vs dictionary that contains orderid and date
          concatenate the datefrom dictionary with orderitem's productid to form a key for a new dictionary
          for this new dictionary's key , have the revenue as the value
          Keep adding the value for the same set of orderdateandprod key.
Step 3: Filter this new dictionary for values containing more than 1000
Step 4: Convert this dictionary into a list of tuples to arrive at the desired output of    
        date,prodid,totalrevenueofthisprodonthisday.

```python
# function to read through a file and split the order lines
def readData(filepath):
    file = open(filepath)
    data = file.read()
    datalist = data.splitlines()
    return datalist
```

```python
orderPath = "/data/retail_db/orders/part-00000"
orders = readData(orderPath)
orderItems = readData("/data/retail_db/order_items/part-00000")
print(orderItems[0:5])
```

```python
def getOrderidanddateindict(orders):
    orderIdanddatedict = {} # create an empty Dictionary
    for i in orders:
        # Dictionary key is the first field of orders record, 
        # Dictionary value is the second field of orders record
        # first field in orders record is Order ID
        # second field in orders record is Date of the order
        orderIdanddatedict[int(i.split(',')[0])] = i.split(',')[1]
    return orderIdanddatedict
```

```python
orderidanddateDict = getOrderidanddateindict(orders)
#orderidanddateDict
```

```python
def getRevenueforProductsoneachDay(orderidanddateDict,orderItems):
    ProductdateandRevenue = {}
    for i in orderItems:
        if(orderidanddateDict.get(int(i.split(',')[1]))):
            #print("orderidfromorderitem",i.split(',')[1])
            productdateandrev = (orderidanddateDict[int(i.split(',')[1])] + '_' + i.split(',')[2],float(i.split(',')[4]))
            #print(productdateandrev[0])
            if(ProductdateandRevenue.get(productdateandrev[0])):
                ProductdateandRevenue[productdateandrev[0]] += productdateandrev[1]
            else:
                ProductdateandRevenue[productdateandrev[0]] = productdateandrev[1]
    return ProductdateandRevenue
    
```

```python
RevenueforProuctoneachDay = getRevenueforProductsoneachDay(orderidanddateDict,orderItems)
```
